Cable transmission line modelling
================

The cable in electrical systems are imposing more challenges to the EMC design as the switching devices running at faster speed. It's even more so in power electronics systems.

Here it proposes an easy method to model the cable transimission line by extracting the lumped sum parameters from performing an easy open circuit (OC) and short circuit (SC) impedance scan using LCR meter.

# Transmission line, open and short circuit

[The input impedance of a transmimssion line](https://resources.system-analysis.cadence.com/blog/msa2021-the-relationship-of-the-input-impedance-of-a-transmission-line-to-load-and-characteristic-impedance) is affected by load and characteristic impedance. 

Define the output terminals with axis point $z=0$ and input terminals $z=-L$. The input impedance is given, 

$$Z_{in}(-L) = Z_0 \dfrac{Z_L + j Z_0 \tan(\beta L)}{Z_0 + j Z_L \tan(\beta L)}$$

When the cable output terminal is open, $Z_L = \infty$

$$Z_{in}^{oc} = -j \dfrac{Z_0}{\tan(\beta L)} = -j Z_0 \cot (\beta L)$$

When the cable output terminal is shorted, $Z_L = 0$

$$Z_{in}^{sc} = j Z_0 \tan (\beta L)$$

We can obtain that, 

$$Z_0 = \sqrt{Z_{in}^{oc} \cdot Z_{in}^{sc}}$$

```python
import numpy as np
import scipy as sp
import pandas as pd

import cmath

import matplotlib as mpl
import matplotlib.pyplot as plt
import glob
from os import path
```

The formulas used below is modified from paper "High frequency modeling for cable and induction motor overvoltage studies in long cable drives". There are actually two ways to obtain the transmission line parameters from open circuit and short circuit test, 

1. More analytical method based on meticulous analysis of dominant parameters in different conditions: high/low frequencies and open/short circuit combinations
2. Obtain $Z_0$ from open and short circuit test, and using matlab to do brute-force regression

```python
f_Rs = lambda Zsc_flo: np.real(Zsc_flo)
f_Ls = lambda fhi, Zsc_fhi: (1/(2*np.pi*fhi)) * np.imag(Zsc_fhi)

f_Rp2 = lambda Zoc_fhi: 1/np.real(1/Zoc_fhi)
f_Cp2 = lambda Zoc_fhi, fhi: np.imag(1/Zoc_fhi)/fhi/2/np.pi

f_Rp1 = lambda Zoc_flo: 1/np.real(1/Zoc_flo)
f_Cp1 = lambda Zoc_flo, flo, Cp2: np.imag(1/Zoc_flo)/flo/2/np.pi - Cp2
```

## Lump sum cable model



### $R_s$: series resistance
At low freq $R_s$ dominate in short circuit test.

```python
Zsc_flo = np.mean(df_sc.Z[0:50])
Rs = f_Rs(Zsc_flo)
Rs
```

### $L_s$: series inductance
At high frequency the inductance dominate in short circuit test.

```python
###### calculate from 500kHz measurement
fhi_sc,zsc_fhi_,deg_fhi_ = df_sc.iloc[(np.abs(df_oc.frequency - 500e3)).idxmin()]
print(fhi_sc,zsc_fhi_,deg_fhi_)

Zsc_fhi = cmath.rect(zsc_fhi_,np.deg2rad(deg_fhi_))
print(Zsc_fhi)

Ls = f_Ls(fhi_sc,Zsc_fhi)
print(Ls)
```

### $R_{p1}$: low frequency resistance

The parallel resistance dominates the low frequency open circuit impedance. 

```python
#### The sampling at 200 Hz looks more reasonable
# can change to other freq to test the results

oc_Rp1 = df_oc.iloc[(np.abs(df_oc.frequency - 200)).idxmin()]

flo_oc = np.array(oc_Rp1.frequency)

Zoc_flo = cmath.rect(oc_Rp1.Z, np.deg2rad(oc_Rp1.phase))
print(Zoc_flo)

Rp1 = f_Rp1(Zoc_flo)
Rp1
```

### $R_{p2}$: parallel resistance 2

At high frequency, $R_{p2}$ and $C_{p2}$ dominates the measured impedance. It could be extracted from the real and imagery part of the measured impedance. 

```python
# Test at 1MHz
oc_Rp2 = df_oc.iloc[(np.abs(df_oc.frequency - 1000e3)).idxmin()]

fhi_oc = np.array(oc_Rp2.frequency)
Zoc_fhi = cmath.rect(oc_Rp2.Z, np.deg2rad(oc_Rp2.phase))

print(oc_Rp2, '\n')
print(Zoc_fhi)

Rp2 = f_Rp2(Zoc_fhi)
Rp2
```

### $C_{p2}$: parallel capacitance 2

```python
Cp2 = f_Cp2(Zoc_fhi, fhi_oc)
Cp2
```

### $C_{p1}$: parallel capacitance 1

```python
Cp1 = f_Cp1(Zoc_flo, flo_oc, Cp2)
Cp1
```

### Cable Surge Impedance $Z_{o}$

The cable characteristic impedance calculated from short- and open-circuit impedances,
$Z_o = \sqrt{Z_{oc} Z_{sc}}$

From Euler's equation, 

$\sqrt{e^{\alpha_1+j\beta_1} \times e^{\alpha_2+j\beta_2}} = \sqrt{e^{\alpha_1+\alpha_2}} \sqrt{e^{j\beta_1+j\beta_2}}  = \sqrt{e^{\alpha_1+\alpha_2}} e^{j\dfrac{\beta_1+\beta_2}{2}}$

```python
Zo = np.sqrt(df_oc.Z * df_sc.Z)
phi_o = (df_oc.phase + df_sc.phase)/2

plt.figure(figsize=(8,6))
ax1 = plt.subplot(2,1,1)
ax1.loglog(df_oc.frequency, Zo)
ax1.grid(which='both')
plt.ylabel('Impedance, ohm')
plt.title('Motor Cable Surge Impedance')

ax2 = plt.subplot(2,1,2)
ax2.semilogx(df_oc.frequency, phi_o)

ax2.grid(which='both')
plt.xlabel('frequency (Hz)')
plt.ylabel('phase, degree')
```

RC Impedance matching Design
==============

Impedance matching is an effective way to minimize reflection and voltage surge. Resistance termination is not feasible for high voltage application due to high losses, thus alternative RC or RLC method could be used. 

The paper "High frequency modeling for cable and induction motor overvoltage studies in long cable drives" proposed the method to calculate the matching network calculation. 

```python
Rx = np.mean(Zo[(df_oc.frequency>3e5) & (df_oc.frequency<1e6)])
Rx
Rx=51
```

```python
Trise = 35e-9
Cx_max = Trise/0.1054/Rx
Cx_min = Trise/0.2231/Rx

print(Cx_max, Cx_min)
```

### $C_{p1}$ capacitance re-calc
According to surge impedance equation

$$|Z_o|_{f\rightarrow\infty} = \sqrt{\dfrac{L_s}{C_{p1}}}$$

```python
Cp1 = Ls/(Rx**2)
Cp1
```

```python
circuit = pd.DataFrame(np.array([[Rs, Ls, Rp1, Cp1, Rp2, Cp2]])*1e6,
             columns=['Rs', 'Ls', 'Rp1', 'Cp1', 'Rp2', 'Cp2'])
circuit
```

```python
len_us = 2.4
len_hv = 2.4
len_jp = 1.8
len_cn = 2.4

Lus_1m = Ls/len_us
Cus_1m = Cp1/len_us
print(Lus_1m, Cus_1m)
```

Voltage ring-up factor, or reflection coefficient $\Gamma$ where $0<\Gamma<1$.

#### Reflection coefficient
$$\Gamma = \dfrac{Z_L-Z_0}{Z_L+Z_0}$$

#### Motor voltage
$$V_{receiving} = (1+\Gamma) V_{sending}$$

#### Cable surge impedance

$$Z_0 = \sqrt{Z_{OC} Z_{SC}}$$

The surge impedance by this method is actually a vector. When the frequency is very high,
$$|Z_o|_{f\rightarrow\infty} = \sqrt{\dfrac{L_s}{C_{p}}}$$

## Variables Affecting Reflected Wave Magnitude
- Cable surge impedance
- Motor surge impedance
- Motor load
- Cable length
- Magnitude of drive pulse (DC link voltage)
- Drive pulse rise time
- Spacing of PWM pulses (responsible for some >2X overvoltage)

```python
trise = np.array([10, 15, 20, 30, 50, 100]) * 1e-9
fe = 1/np.pi/trise # effective switching frequency
c = 3e8
v= 1/np.sqrt(Lus_1m*Cus_1m) # travelling wave velocity
wavelength = v/fe
lc = wavelength/4

# print(fe, '\n', lc)

plt.figure(figsize=(8,6))
for i_, lc_ in enumerate(lc):

    l_ = np.linspace(0.1, lc_, 50)

    Vpk = np.sin(np.pi/2 * (l_/lc_)) + 1
    plt.semilogx(l_, Vpk,
                 label='{:.0f} ns, 2pu Vpk length {:.1f}m'.format(trise[i_]*1e9, lc_),
                 linewidth=3)

plt.grid(which='both')
plt.legend(title='drive voltage rising time')
plt.xlabel('Cable length (m)')
plt.ylabel('normalized over-voltage (Vpk/Vbus, pu)')
plt.title('cable over-voltage estimation @ speed ={:.2f}c'.format(v/c))
```
