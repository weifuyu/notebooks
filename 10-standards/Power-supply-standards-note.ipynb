Source: [DOE Appliance Standards for External Power Supplies](https://www.power.com/community/green-room/blog/doe-appliance-standards-external-power-supplies-2023)

# 80 PLUS Certificate for Internal Power Supplies (IPS) 
Source: [80 PLUS](https://en.wikipedia.org/wiki/80_Plus)

80 Plus (trademarked 80 PLUS) is a voluntary certification program launched in 2004, intended to promote efficient energy use in computer (internal) power supply units (PSUs).
